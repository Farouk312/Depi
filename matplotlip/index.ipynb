```python
import matplotlib
print(matplotlib.__version__)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('tips')
sns.lineplot(data=df,x='total_bill',y='tip')
plt.title('line plot')
plt.show()
```

```python
df.info()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.title('scatter plot')
plt.show()
```

```python
sns.lmplot(data=df,x='total_bill',y='tip')
plt.title('lm plot')
plt.show()
```

```python
df['total_bill'].nunique()
```

```python
sns.histplot(data=df,x='total_bill',bins=20)
plt.title('histogram')
plt.show()
```

```python
#not best way to determine the number of bins but it is one of the ways
sns.histplot(data=df,x='total_bill',bins=229)
plt.title('histogram')
plt.show()
```

```python

sns.histplot(data=df,x='total_bill',bins='auto')
plt.title('histogram')
plt.show()
```

```python
sns.barplot(data=df,x='day',y='total_bill')
plt.title('bar plot')
```

```python
sns.pairplot(data=df)  #research about it
plt.show()
```

```python
sns.violinplot(data=df,x='day',y='total_bill')
plt.title('violin plot')
plt.show()
```
