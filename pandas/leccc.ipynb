```python
import pandas as pd
```

```python
df=pd.read_csv('births.csv')
df.info()
```

```python
import numpy as np
```

```python
df['decade'] = 10 * (df['year'] // 10)
 
```

```python
df.info()
```

```python
df.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
```

```python
import matplotlib.pyplot as plt

df.pivot_table('births', index='decade', columns='gender', aggfunc='sum').plot()   #plot(kind='bar')
plt.ylabel('total births per decade')

```

```python

df.pivot_table('births', index='decade', columns='gender', aggfunc='sum').plot(kind='bar')
plt.ylabel('total births per decade')
```
