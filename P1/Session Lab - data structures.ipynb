### Dictionaries
A dictionary in Python is a collection of key-value pairs. Each key is connected
to a value, and you can use a key to access the value associated with that key.


A keyâ€™s value can be a number, a string, a list, or even another dictionary.


```python
user = {'Name': 'Ali', 'Age': 25}  
        

#print(user['Name'])
#print(user['Age'])

user['phone']='010000000'
print(user)
```

```python
user2 = {10 : 20, 30:5}
user2
```

```python
print(user2[30])
```

```python
# Adding New Key-Value Pairs

user = {'Name': 'Ali', 'Age': 25}

user['Address'] = 'Egypt'
user['weight'] =  70
print(user)
```

```python
user['Ali']
```

```python
# Modifying Values in a Dictionary
user['Address'] = 'USA'
```

```python
user
```

```python
# Removing Key-Value Pairs
del user['Address']
user 
```

```python
# A Dictionary of Similar Objects
favorite_languages = {
'mohamed': 'python',
'sara': 'c',
'emad': 'ruby',
'sara': 'c++',
}
#print(favorite_languages['sara'])
print(favorite_languages['ola'])
```

```python
favorite_languages
```

```python
#dictionary.get(key, default_value)

x = 'max'
ola_value = favorite_languages.get(x, f'No {x} value assigned.')
print(ola_value)
```

```python
user = [{'Name': 'Ali', 'Age': 25},{'Name':'Beshoy','Age':22}]
print(user)
```

```python
#Looping Through a Dictionary
for a, b in user.items():
    print(f"\nKey: {a}")
    print(f"Value: {b}")

    
#dictionaryname.items()   ---> all the keys and values
#dicname.keys()   ---> All keys
#dicname.values()---> all values
    
```
