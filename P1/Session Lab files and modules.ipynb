### File Handling

The `open()` function is used to open a file for reading, writing, or appending.
- Returns a file object that can be used for file operations.
- Takes a filename and a mode as its arguments.
The function also takes the mode for opening the file:
- `'r'` opens the file in read only mode.
- `'w'` to write data to a file.


```python
with open('done/data.txt', 'r') as f:
    data = f.read()
```

```python
with open('done/data.txt', 'w') as f:
    data = 'some data to be written to the file'
    f.write(data)
```

**File Opening Modes**

Different modes define how a file is opened:

| Mode | Description |
|------|------------|
| `"r"` | Read mode (default), file must exist |
| `"w"` | Write mode, creates or overwrites file |
| `"a"` | Append mode, adds to existing file |
| `"r+"` | Read and write mode |


```python
with open("done/data.txt", "a") as f: # Opens file in append mode
    f.write("\nAppending new content.")  # Adds text without erasing existing content
```

The `readlines()` function reads a file line by line and returns a list of strings.

```python
with open("done/file.txt", "r") as f:
    lines = f.readlines()  # Reads file into a list of lines
for line in lines:
    print(line.strip())  # Removes newline characters before printing
```

### Exceptions and Error Handling

Using `try` and `except` blocks for error handling, where the try block is used to wrap code that might raise an exception, and the except block handles the error gracefully.

```python
try:
    print(5 / 0)  # This raises ZeroDivisionError
except ZeroDivisionError:
    print("You can't divide by zero!")
```

You can use the `else` block for code that runs when no exception occurs.

```python
try:
    answer = 10 / 2  # No error
except ZeroDivisionError:
    print("You can't divide by zero!")
else:
    print(f"The result is {answer}")
```

Sometimes, we donâ€™t need to notify the user about every exception. The `pass` keyword allows the program to fail silently.

```python
try:
    with open("non_existent_file.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    pass  # The program will continue without errors
```

**Handling Multiple Exceptions:**

A try block can catch multiple exceptions using separate `except` blocks.


```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("You can't divide by zero!")
```

**Raising Custom Exceptions:**

You can define custom errors using the raise keyword.

```python
def withdraw(amount):
    if amount > 1000:
        raise ValueError("Withdrawal limit exceeded!")  # Custom error message
    print(f"Withdrawing {amount} dollars")

try:
    withdraw(1500)
except ValueError as e:
    print(f"Transaction failed: {e}")
```
